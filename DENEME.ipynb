```python
# KelimeOYUNU

from random import choice


adamCan = 4
kelimeler = ["mürekkep", "dolma kalem", "lili", "sakarya"]
secilenKelime = choice(kelimeler)
#print(secilenKelime)

while adamCan > 0:
    tahmin = input("Kelimeyi tahmin ediniz: ")
    if tahmin == secilenKelime:
      print("Tebrikler :)")
      break
    else:
      adamCan-=1
      print(str(adamCan) + " canınız kaldı")
```
